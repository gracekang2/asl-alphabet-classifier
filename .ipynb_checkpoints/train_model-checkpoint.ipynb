```python
import torch
import torchvision
from torchvision import datasets, transforms
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import pandas as pd
```

```python
train_data = pd.read_csv('/storage/asl/sign_mnist_train.csv')
test_data = pd.read_csv('/storage/asl/sign_mnist_test.csv')

# img_dataset = datasets.ImageFolder('/storage/asl/images', transform=img_transform)
# N = len(img_dataset)
# train_data, val_data = torch.utils.data.random_split(img_dataset, lengths=[int(N*0.8), int(N*0.2)])
# trainloader = torch.utils.data.DataLoader(train_data, batch_size=32, shuffle=True)
# validateloader = torch.utils.data.DataLoader(val_data, batch_size=32, shuffle=True)
```

```python
class MNISTDataSet(Dataset):
    def __init__(self, data, transform = None):
        """
        :data type: Pandas df
        :transform: callable, optional
        """
        self.data = data
        self.transform = transform
        
    def __len__(self):
        return self.data.shape[0]
    
    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        label = row[0]
        if label >= 10:
            label -= 1
        return (label, torch.reshape(torch.Tensor(row[1:]), (1, 28, 28)))
    
img_transform = transforms.Compose([
    # transforms.Resize(128),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
    # transforms.Grayscale(),
])
    
train_dataset = MNISTDataSet(train_data, transform=img_transform)
validate_dataset = MNISTDataSet(test_data, transform=img_transform)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(npimg, cmap='gray')
    plt.show()

# images, labels = next(iter(trainloader))
# imshow(images[0])
# print(labels[0])

sample = train_dataset[0]
print(sample[0], sample[1].shape)

imshow(sample[1][0])
# print(sample[1][0])
```

```python
trainloader = DataLoader(train_dataset, batch_size=64, shuffle=True)
validateloader = DataLoader(validate_dataset, batch_size=64, shuffle=True)

bx,by = next(iter(trainloader))
bx
```

```python
by.shape
```

```python
class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=(5,5))
        self.conv2 = nn.Conv2d(32, 64, kernel_size=(3,3))
        self.dropout1 = nn.Dropout2d(0.25)
        self.dropout2 = nn.Dropout2d(0.25)
        self.dropout3 = nn.Dropout2d(0.5)
        self.fc1 = nn.Linear(1600, 256)
        self.fc2 = nn.Linear(256, 24)
        
    def forward(self, x):
        x = self.conv1(x)
        x = F.relu(x)
        x = F.max_pool2d(x, kernel_size=(2,2))
        x = self.dropout1(x)

        x = self.conv2(x)
        x = F.relu(x)
        x = F.max_pool2d(x, kernel_size=(2,2), stride=(2,2))
        x = self.dropout2(x)
        
        x = torch.flatten(x, 1)
        x = self.fc1(x)
        x = F.relu(x)
        x = self.dropout3(x)
        
        x = self.fc2(x)

        return x

net = NeuralNet()
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, weight_decay=0.1)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=2, verbose=True, min_lr=0.00001, factor=0.5)
```

```python
EPOCHS = 10
N = len(trainloader)

losses = []
accuracies = []
len(trainloader)
for epoch in range(EPOCHS):
    print("EPOCH %d" % (epoch + 1))
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        labels, inputs = data
        optimizer.zero_grad()
        preds = net(inputs)
        loss = criterion(preds, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss
      
    epoch_loss = running_loss / N
    print('[%d, %5d] loss: %.3f' %
        (epoch + 1, i + 1, epoch_loss))
    losses.append(epoch_loss)
    running_loss = 0.0
    
    correct = 0
    total = 0
    accuracy = 0
    with torch.no_grad():
        for data in validateloader:
            labels, images = data
            outputs = net(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
        accuracy = (correct/total) * 100
        scheduler.step(metrics=accuracy)
        print('Accuracy: %d %%' % (accuracy))
        accuracies.append(accuracy)

print('Finished')
plt.plot(losses)
plt.plot(accuracies)
plt.show()
```
